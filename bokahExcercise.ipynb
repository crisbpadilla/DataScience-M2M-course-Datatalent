```python
from bokeh.plotting import figure,show
from bokeh.io import output_notebook

output_notebook()
```

```python
import pandas as pd

cld=pd.read_csv('ClassData.csv')
cld.info()
print(cld)
# in bokeh the to create a bar chart it is necesary consider that the data in x
# axis have to be a categorical value, wich means a string
x=cld['Grade'].apply(str)
y=cld['Total']
```

```python
# also bokeh different from matplotlib is that they use a source, which is a
# dictionary with the variables used in the different axis, so it is necesary to
# create it before run the visualization
source={'x':x,'y':y}
# first step to build up the graphic is declaring the components that will be
# shown in the graph
visual = figure(title='Total enrollment per Grade',x_range=x,y_range=(0,40),
              x_axis_label='Grades',y_axis_label='Total enrollment',
              height=300,width=800)
# here I declare my key arguments that will give the specific characteristics of
# my graph

visual.vbar(x='x',top='y',source=source, width= 0.4,legend_field='x')
visual.legend.orientation= 'horizontal'
visual.legend.location= 'top_center'
visual.xgrid.grid_line_color=None

show(visual)
```

```python
from bokeh.transform import factor_cmap
from bokeh.palettes import Category20, Viridis
import random


a=cld['Grade'].apply(str)
b=cld['Female']

```

```python
#this library is used to color the chart, in this case the bars
source={'a':a,'b':b}
# first step to build up the graphic is declaring the components that will be
# shown in the graph
visual_color = figure(title='Enrollment of Woman per Grade',x_range=a,y_range=(0,30),
              x_axis_label='Grades',y_axis_label='Female enrollment',
              height=300,width=800)
# here I declare my key arguments that will give the specific characteristics of
# my graph

visual_color.vbar(x='a',top='b',source=source, width= 0.6,legend_field='a',
            fill_color= factor_cmap('a',palette = (Category20[cld['Grade'].count()]),factors=a))

visual_color.legend.orientation= 'horizontal'
visual_color.legend.location= 'top_center'
visual_color.xgrid.grid_line_color=None

show(visual_color)
```

## Time series charts

Create a time series chart with the stock volume for tesla

```python
import pandas as pd

tsla= pd.read_csv('TSLA.csv')
tsla['date']= pd.to_datetime(tsla['date'])
tsla.head()
```

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import NumeralTickFormatter

output_notebook()

date = tsla['date']
volume= tsla['volume']

base= figure(title='Volume Stock tesla', x_axis_label='Dates', y_axis_label='Volume', x_axis_type='datetime')

base.line(x=date,y=volume,color='#FF0000')
# this allows to show the Y values in a milesimal format, using the 0.0a to use
# a suffixes as M, k , etc. depending the lenght of the value
base.yaxis.formatter = NumeralTickFormatter(format="0,0a")

show(base)
```

```python
import pandas as pd
from bokeh.plotting import show,figure
from bokeh.io import output_notebook


output_notebook()

stock_files=['AAPL','AMZN','GOOG','MSFT','TSLA']
dic_file={}
for files in stock_files:
   dic_file[files]=pd.read_csv(f'{files}.csv')

lst_apl= dic_file['AAPL'][-90:]
lst_amz= dic_file['AMZN'][-90:]
lst_gog= dic_file['GOOG'][-90:]
lst_msf= dic_file['MSFT'][-90:]
lst_tla= dic_file['TSLA'][-90:]

print(lst_tla)
```

```python
def ptg_varstock(data):
  data=list(data)
  return [((data[i]-data[0])/data[0])*100 for i in range(len(data))]

print(ptg_varstock(lst_tla['open']))
```

```python
from bokeh.layouts import row,column

y_apl= ptg_varstock(lst_apl['close'])
y_amz= ptg_varstock(lst_amz['close'])
y_gog= ptg_varstock(lst_gog['close'])
y_msf= ptg_varstock(lst_msf['close'])
y_tla= ptg_varstock(lst_tla['close'])
x= pd.to_datetime(lst_amz['date'])
```

```python
stocks_visual= figure(title='Stock variation by %', x_axis_label='Dates',
                      y_axis_label='Stock variation',x_axis_type='datetime')
stocks_visual.line(x=x,y=y_apl,color='#ff0000',legend_label='AAPL')
stocks_visual.line(x=x,y=y_amz,color='#E53D00',legend_label='AMZN')
stocks_visual.line(x=x,y=y_gog,color='#202C59',legend_label='GOOG')
stocks_visual.line(x=x,y=y_msf,color='#F8C630',legend_label='AMZN')
stocks_visual.line(x=x,y=y_tla,color='#046865',legend_label='TSLA')

stocks_visual.legend.location='top_left'
show(stocks_visual)
```
