## Cleaning data

In this part of the course, we will use methods from pandas library to detect data that is not numerical (NaN 'Not a number') and erase it or replace it regarding the scenario.


```python
import pandas as pd

df1=pd.read_csv('MSFT.csv')
#first change date data type from object to datatime
df1.dtypes
```

```python
#now we will obtain all the non numerical values
columns_check=['high','low','close']
df1[columns_check]=df1[columns_check].apply(pd.to_numeric, errors='coerce')
df1['date']= pd.to_datetime(df1['date'])
print(df1)
#so there is no missing values within this dataframe

#nowe we will replace the index label to the date
df1=df1.drop(['volume','open'],axis=1)
df1.set_index('date', inplace=True)
df1.head()
```

## Calculations

Creting new column with calculation between numerical values of other columns.

* Get the difference bettwen high and low columns
* Return the higher value between open and close columns
* return the close price as a percentaje of the close price.

```python
#we will work with just the table of AMZN
amzn['stock_var']= amzn['high']-amzn['low']
#to obtain the higher value between open and close we will use the method max
amzn['max_val']= amzn[['open','close']].max(axis=1)
amzn.head()

# finally to return the percentage, we create a constant with the first value
# that will represent the 100% and then calculate the percentage of the future
# values

cpp=amzn.loc[0,'close']
Per_change=(amzn['close']/cpp)*100
amzn['% close_change']=Per_change

amzn.head()
```

## Functions with pandas

As happened with the arrays, pandas also has a  method to applying more complex functions along the rows of a dataframe.

  So as part of the excercises, we are going to:

  * Create a new volume_estimate by applying a function to the volumn column wich replaces values grater than 10.000.000 with high and everything else with low.

```python
def vol_val(data):
  if data['volume']>10000000:
    return 'high'
  else:
    return 'low'

amzn['volume_estimate']= amzn.apply(vol_val, axis=1)
amzn.head()
```
